```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from wordcloud import STOPWORDS, ImageColorGenerator, WordCloud
```

```python
mask = np.array(Image.open("./assets/wordcloud/side_airplane_02w.png"))
```

```python
np.unique(
    mask, return_index=False, return_inverse=False, return_counts=False, axis=None
)
```

```python
# Content
text = open("./assets/wordcloud/regional_airport_wordcloud.txt", "r").read()
stopwords = set(STOPWORDS)

# Appearance
wc = WordCloud(
    background_color="white",
    stopwords=stopwords,
    mask=mask,
    contour_width=1,
    contour_color="lightgray",
    max_words=2000,
    max_font_size=40,
    random_state=42,
)

# generate word cloud
wc.generate(text)

# create coloring from image
image_colors = ImageColorGenerator(mask)

# wc.recolor(color_func=image_colors)

wc.to_file('./assets/wordcloud/regional_airport_wordcloud.png')

plt.figure(figsize=(15, 10))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```
