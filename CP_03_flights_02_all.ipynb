```python
# loading libraries

import glob
import os
import zipfile

import folium
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
import requests
import seaborn as sns
import wget

# web scrapping with BeautifulSoup
from bs4 import BeautifulSoup
```

```python
# web scrapping with BeautifulSoup – links to download flight data from ANAC
```

```python
url = "https://www.gov.br/anac/pt-br/assuntos/regulados/empresas-aereas/envio-de-informacoes/microdados/microdados"
html = requests.get(url)
soup = BeautifulSoup(html.text, "html.parser")

all_links = [x.get('href') for x in soup.find_all('a', href=True)] 
flight_data_links = []

for x in all_links:
    if 'envio-de-informacoes/microdados/basica' in x:
        flight_data_links.append(x)
```

```python
flight_data_links[0]
```

```python
os.getcwd()
```

```python
# foltder to save the files
os.chdir('')

# download files
for links_data in flight_data_links:
    wget.download(f'{links_data}')
```

```python
# unzip files

extension = ".zip"

for item in os.listdir():
    if item.endswith(extension): # check for ".zip" extension
        file_name = os.path.abspath(item) # get full path of files
        zip_ref = zipfile.ZipFile(file_name) # create zipfile object
        zip_ref.extractall() # extract file to dir
        zip_ref.close() # close file
        os.remove(file_name) # delete zipped file
```
