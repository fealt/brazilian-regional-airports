```python
# loading libraries

import folium
from folium.plugins import FastMarkerCluster
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
import seaborn as sns

import io
from PIL import Image
```

```python
pd.set_option("display.max_columns", None)
```

```python
df_2002_2021s = pd.read_csv("./assets/csv/df_clusters.csv", sep=",")
```

```python
count_perc = df_2002_2021s.groupby('cluster')[['count', 'percentage']].sum().reset_index()
```

```python
count_perc
```

```python
# Set up with a higher resolution screen (useful on Mac)
%config InlineBackend.figure_format = 'retina'

plt.figure(figsize=(8, 8))
plt.title('Clusters and %')
# Plot DataFrame
sns.barplot(x='cluster',y='percentage', 
           data=count_perc, color="c")
plt.show()
```

```python
map_clusters = folium.Map(location=[-15, -55], tiles='Cartodb Positron', zoom_start=4)
n_mean = df_2002_2021s['count'].mean()

# set color scheme for the clusters
x = np.arange(8)
ys = [i + x + (i * x) ** 2 for i in range(8)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lng, number, cluster, city, state in zip(
    df_2002_2021s["lat_dd"],
    df_2002_2021s["long_dd"],
    df_2002_2021s["count"],
    df_2002_2021s["cluster"],
    df_2002_2021s["city_primary"],
    df_2002_2021s["state"],
):
    # label = folium.Popup(str(city)+ ','+str(state) + '- Cluster ' + str(cluster), parse_html=True)
    folium.vector_layers.CircleMarker(
        [lat, lng],
        radius=number/n_mean*2,
        # popup=label,
        tooltip=str(city) + ", " + str(state) + " – Cluster " + str(cluster),
        color=rainbow[cluster - 1],
        fill=True,
        fill_color=rainbow[cluster - 1],
        fill_opacity=0.2,
    ).add_to(map_clusters)

# map_clusters.save('map_clusters.html')

map_clusters
```
